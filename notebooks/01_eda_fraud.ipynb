# Fraud dataset – EDA

Quick look at the credit card fraud dataset: class imbalance, basic stats for `Amount` and `Time`, and a couple of simple plots.

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

base = Path.cwd().parent
csv_path = base / "data" / "creditcard.csv"
df = pd.read_csv(csv_path)
df.head()
```

Class distribution (0 = normal, 1 = fraud).

```python
counts = df["Class"].value_counts()
print(counts)
print((counts / len(df)).round(6))
sns.countplot(x="Class", data=df)
plt.title("Class distribution")
plt.show()
```

Basic stats for `Amount` and `Time`, split by class.

```python
df.groupby("Class")["Amount"].describe()
```

```python
sns.boxplot(x="Class", y="Amount", data=df)
plt.title("Amount by class")
plt.show()
```

Scatter of `Amount` vs `Time` (subsampled for speed), colored by class.

```python
sample = df.sample(frac=0.05, random_state=42)
plt.figure(figsize=(6, 4))
sns.scatterplot(data=sample, x="Time", y="Amount", hue="Class", alpha=0.4)
plt.title("Amount vs Time (sample)")
plt.show()
```

Quick takeaways you can mention in a README:
- Fraudulent transactions are a tiny fraction of all rows.
- Amount distribution for fraud vs non-fraud can be quite different (depending on the dataset).
- Any model must be evaluated with metrics that focus on the fraud class (recall/precision), not just accuracy.
